```python
import obspy
import matplotlib.pyplot as plt
from obspy.imaging.beachball import beachball

# Load the moment tensor parameters from the XML file
# Replace 'moment_tensor.xml' with the actual path to your XML file
moment_tensor = obspy.read_events('quakeml.xml')[0]

# Extract moment tensor parameters
moment_tensor_m_rr = moment_tensor.m_rr
moment_tensor_m_tt = moment_tensor.m_tt
moment_tensor_m_pp = moment_tensor.m_pp
moment_tensor_m_rt = moment_tensor.m_rt
moment_tensor_m_rp = moment_tensor.m_rp
moment_tensor_m_tp = moment_tensor.m_tp

# Create a beachball diagram
mt_params = [moment_tensor_m_rr, moment_tensor_m_tt, moment_tensor_m_pp,
             moment_tensor_m_rt, moment_tensor_m_rp, moment_tensor_m_tp]
beachball(mt_params, size=200, linewidth=2, facecolor='b')

# Customize the plot (optional)
plt.title('Moment Tensor Focal Mechanism')
plt.gca().set_aspect('equal')
plt.grid(True)

# Show the plot or save it to a file
plt.show()
```
